```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
```

```python

age_counts = df['Age'].dropna().astype(int).value_counts().sort_index()
plt.plot(age_counts.index, age_counts.values)
plt.title('Passenger Count per Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python

df['Age'].hist(bins=30, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python

gender_counts = df['Sex'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Gender Distribution')
plt.axis('equal')
plt.show()
```

```python

fare_bins = [0, 10, 20, 50, df['Fare'].max()]
fare_labels = ['0-10', '10-20', '20-50', '50+']
fare_categories = pd.cut(df['Fare'], bins=fare_bins, labels=fare_labels, include_lowest=True)
fare_counts = fare_categories.value_counts().sort_index()
plt.pie(fare_counts, labels=fare_labels, autopct='%1.1f%%', startangle=90)
plt.title('Fare Category Distribution')
plt.axis('equal')
plt.show()
```

```python

df.boxplot(column='Age', by='Pclass')
plt.title('Age vs Passenger Class')
plt.suptitle('')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.tight_layout()
plt.show()
```
